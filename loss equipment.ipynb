```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
DataFrame=pd.read_csv("../input/2022-ukraine-russian-war/russia_losses_equipment.csv", index_col="date", parse_dates=True)
DataFrame
```

```python
DataFrame.describe()
```

let's make a visualization to see the losses the Russian War Machine has incurred over time. Since we are using time as a variable, it would be a good idea to use a line graph. Also, to avoid some potential issues, I am going to remove the "day" attribute and work only with the "date" index.

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20,9))
plt.title("Russian Equipment Losses")
plt.xlabel("Date")
plt.ylabel("Asset")
sns.lineplot(data=DataFrame.drop("day", axis=1))
plt.show()
```

There are three bins that trend positively: fuel, tanks, and military auto. It could also be argued that fuel expenditures have a relationship with tank and military auto losses. Let's take a closer look at these three variables.

```python
plt.figure(figsize=(20,9))

x_data = ["military auto", "APC"]

for vehicle in x_data:
    sns.lmplot(data=DataFrame, x=vehicle, y="tank")
plt.show()
```

Military auto losses have a large, positive trending relationship. At the static rate of 60 fuel tanks lost per day, this behavior is to be expected. The more fuel that is lost, the more likely it is to be that vehicle losses will incure due to lack of fuel and/or Ukranian acquisition.

```python
plt.figure(figsize=(7,7))
DataFrame.sum().sort_values().plot(kind='pie',autopct=lambda p:'{:.2f}%\n({:.0f})'.format(p,(p/100)*(DataFrame.sum().sort_values().sum())))
plt.title('Destroyed Equepments')
```
